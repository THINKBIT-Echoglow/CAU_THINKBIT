아래 코드에서 cv2.cvtColor = convert color의 의미로 이미지 색상변경을 위한 함수 사용

(image - 우리가 imread를 통해 읽어온 이미지를 저장한 변수)

(cv2.COLOR_BGR2GRAY - image는 BGR 3색의 조합으로 구성된 이미지이므로, 이를 2 = to GRAY 흑백 색상으로 변환)

(저장하고자 하는 경로 및 파일명, 저장할 이미지 - 변수)

```python
import numpy as np

image = cv2.imread(image_path)

# 이미지 크기 확인
height, width, channels = image.shape

# 가우시안 노이즈 생성
mean = 0
std_dev = 25
gaussian_noise = np.random.normal(mean, std_dev, (height, width, channels)).astype(np.uint8)

# 원본 이미지와 노이즈를 더함
noisy_image = cv2.add(image, gaussian_noise)

# 결과 이미지 출력
cv2_imshow(noisy_image)
```

```python
image = cv2.imread(image_path)

# 이미지 크기 확인
height, width, channels = image.shape

# 가우시안 노이즈 생성
mean = 0
std_dev = 25
gaussian_noise = np.random.normal(mean, std_dev, (height, width, channels)).astype(np.uint8)

# 원본 이미지와 노이즈를 더함
noisy_image = cv2.add(image, gaussian_noise)

# 가우시안 블러 적용
blurred_image = cv2.GaussianBlur(noisy_image, (15, 15), 0)

# 결과 이미지 출력
cv2_imshow(blurred_image)
```

```python
import numpy as np

image = cv2.imread(image_path)
original_image = image.copy()  # 원본 이미지 복사

# 이미지를 그레이스케일로 변환
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 가우시안 블러 적용 (노이즈 제거)
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# 케니 엣지 검출기 적용
edges = cv2.Canny(blurred, 50, 150)

# 윤곽선 검출
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 모든 윤곽선을 그리기 위한 빈 이미지 생성
contour_image = np.zeros_like(image)

# 모든 윤곽선을 그리기
cv2.drawContours(contour_image, contours, -1, (255, 255, 255), 1)  # 세 번째 인자는 윤곽선 두께

# 결과 이미지 출력
cv2_imshow(contour_image)
```
