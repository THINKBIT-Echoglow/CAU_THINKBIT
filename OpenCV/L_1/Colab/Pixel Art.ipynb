```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image_path = '/content/drive/MyDrive/L_1/flower.jpg'
image = cv2.imread(image_path)
```

```python
image
```

```python
if image is None:
    print(f'오류: {image_path}에서 이미지를 로드할 수 없습니다.')
else:

    small_image = cv2.resize(image, None, fx=0.1, fy=0.1, interpolation=cv2.INTER_NEAREST)

    pixel_art_image = cv2.resize(small_image, (image.shape[1], image.shape[0]), interpolation=cv2.INTER_NEAREST)

    cv2_imshow(pixel_art_image)
```

```python
image_data = np.array([
    [[255, 0, 0], [0, 255, 0]],
    [[0, 0, 255], [255, 255, 0]]
], dtype=np.uint8)

# 이미지 데이터를 이미지로 변환
image = cv2.cvtColor(image_data, cv2.COLOR_RGB2BGR)

# 이미지 출력
cv2_imshow(image)

save_path = '/content/drive/MyDrive/L_1/2x2.png'
cv2.imwrite(save_path, image)
```
