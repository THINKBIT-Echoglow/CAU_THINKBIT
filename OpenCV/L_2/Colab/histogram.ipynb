```python
#OPENCV (=cv2)를 import하여 사용한다.
# 구글코랩에서는 cv2.imshow기능이 사용불가하여, 추가적인 프로세스를 거쳐 진행한다.
import cv2
from google.colab.patches import cv2_imshow
```

```python
#이미지의 경로 지정 - 이미지파일 - 우클릭 - 경로복사
image_path = '/content/drive/MyDrive/L_2/Colab/river.jpg'
```

```python
#이미지  출력
image = cv2.imread(image_path)
cv2_imshow(image)
```

```python
import matplotlib.pyplot as plt
def plot_gray_histogram(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # 히스토그램 계산
    histr = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

    # 히스토그램 그리기
    plt.plot(histr, color='gray')
    plt.title('Grayscale Histogram')
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.xlim([0, 256])
    plt.show()
```

```python
plot_gray_histogram(image)
```

```python
def plot_histogram(image, title, color='gray'):
    # 히스토그램 계산
    histr = cv2.calcHist([image], [0], None, [256], [0, 256])

    # 히스토그램 그리기
    plt.plot(histr, color=color)
    plt.title(title)
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.xlim([0, 256])
    plt.show()

def equalize_histogram(image):
    # 이미지를 흑백으로 변환
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # 히스토그램 평활화
    equalized_image = cv2.equalizeHist(gray_image)

    return gray_image, equalized_image

# 원본 이미지와 평활화된 이미지 생성
gray_image, equalized_image = equalize_histogram(image)

# 원본 흑백 이미지 히스토그램
plot_histogram(gray_image, 'Original Grayscale Histogram')

# 평활화된 흑백 이미지 히스토그램
plot_histogram(equalized_image, 'Equalized Grayscale Histogram')

# 원본 이미지와 평활화된 이미지 보여주기
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.title('Original Grayscale Image')
plt.imshow(gray_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Equalized Grayscale Image')
plt.imshow(equalized_image, cmap='gray')
plt.axis('off')

plt.show()
```
